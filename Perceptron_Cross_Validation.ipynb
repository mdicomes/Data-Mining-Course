```python
import numpy as np 
import pandas as pd
```

```python
# Open .txt file using appropriate delimiter
df_data = pd.read_csv('dataForAssignment2.txt', delimiter=',', header=None)

# Clean up the extra spaces and then split the columns
df_data_clean = df_data[0].str.strip().replace(r'\s+', ' ', regex=True)

# SPlit the columns using a single space as the delimiter
df_data_split = df_data_clean.str.split(' ', expand=True)

# Convert to numeric, coercing any non-convertible values to NaN
df_data_numeric = df_data_split.apply(pd.to_numeric, errors='coerce')
# Make a copy of the original to use in dimensionality reduction
df_data_numeric_copy = df_data_numeric.copy()

print("Shape of dataset: ", df_data_numeric.shape)
print(df_data_numeric.head())
```

```python
## Normalizing Dataset ##
# Z-normalize across ROWS
df_norm_data_original = df_data_numeric.apply(lambda x: (x - x.mean()) / (x.std()), axis=1) # axis = 1 indicates row
print(df_norm_data_original.shape)
print(df_norm_data_original.head())
```

```python
## Add label column ##

# Define number of rows #
n_rows = df_norm_data_original.shape[0]

# Create the new column: first 8000 rows are '0', last 8000 rows are '1' #
df_norm_data_original['label'] = [0]*(n_rows // 2) + [1]*(n_rows - n_rows // 2)

print("Shape of dataset with label column: ", df_norm_data_original.shape)
print(df_norm_data_original.head(5))
```

```python
## Shuffle the dataset ##
df_data_shuffle = df_norm_data_original.sample(frac=1, random_state = 1234).reset_index(drop=True)

print(df_data_shuffle.head(10))
```

```python
##---------------- 5 FOLD CROSS VALIDATION SPLITS ------------------------------##
## Splitting the dataframe into 5 parts (20% of total rows per split) ##
# number of splits #
n_splits = 5
split_size = len(df_data_shuffle) // n_splits # number of rows per split

# List of splits #
splits = []

for i in range(n_splits):
    start_idx = i * split_size # indicates where each split should start: first split start_idx = 0, second split start_idx = split_size * 1 (i = 1), third split start_idx = split_size * 2 (i = 2), etc.
    if i != n_splits - 1:
        end_idx = (i+1) * split_size
    else:
        end_idx = len(df_data_shuffle) # last row will take remaining rows

    split = df_data_shuffle.iloc[start_idx:end_idx] # creating each fold
    splits.append(split)

print("Length of splits list: ", len(splits))


##----------------------- CROSS-VALIDATION ------------------------------##

accuracies = []
iterations_per_fold = []

for fold in range(n_splits):
    D_i = splits[fold] # this is the test set
    train_set = pd.concat([splits[j] for j in range(n_splits) if j != fold])

    # Output the result for this split
    print(f"Fold {fold + 1}:")
    print(f"Train set size: {len(train_set)}, Test set size: {len(D_i)}")
    #print("Test set sample: \n", D_i.head(), "\n")
    
    # Split data from labels in order to calculate accuracies per split
    X_train = train_set.drop("label", axis=1)
    y_train = train_set["label"]
    X_test = D_i.drop("label", axis=1)
    y_test = D_i["label"]

    ## Create a weight matrix and add a column of 1s to the dataset ##
    samples, features = X_train.shape

    # Add the column of 1s to match bias term ##
    X_train_ones = np.hstack([np.ones((samples, 1)), X_train])
    X_test_ones = np.hstack([np.ones((len(X_test), 1)), X_test])
    # print("Shape of the new training set: ", X_train_ones.shape)
    # print("Shape of the new test set: ", X_test_ones.shape)

    # CREATE AND INITIALIZE WEIGHT VECTOR #
    weight_mat = np.zeros(features + 1) # initializing with zeros including the w0 column
    # print("Shape of the weight matrix: ", weight_mat.shape)

    ## Training the perceptron ##
    r = 0.01 # learning rate
    max_epochs = 1000
    converged = False
    epoch = 0

    for epoch in range(max_epochs):
        weight_updates = 0
        for i in range(samples):
            # Compute weighted sum (z = W.X)
            z = np.dot(X_train_ones[i], weight_mat)

            # Activation function: step function
            y_pred = 1 if z > 0 else 0

            # Calculate error
            error = y_train.iloc[i] - y_pred

            # Update weights based on errors
            if error != 0:
                weight_mat += r * error * X_train_ones[i] # weight update: weight = weight + learning rate * (expected - predicted) * x
                weight_updates += 1

        # If no weight updates, convergence is reached
        if weight_updates == 0:
            converged = True
            break # early stopping

    # Store the number of iterations (epochs) for this fold
    iterations_per_fold.append(epoch + 1)

    ##------------------- EVALUATE THE MODEL --------------------##
    correct_predictions = 0
    for i in range(len(X_test_ones)):
        # Compute weighted sum
        z = np.dot(X_test_ones[i], weight_mat)
        y_pred = 1 if z > 0 else 0

        # Count correct predictions
        if y_pred == y_test.iloc[i]:
            correct_predictions += 1
        
    # Calculating accuracies per fold
    accuracy = correct_predictions / len(X_test_ones)
    accuracies.append(accuracy)
    print(f"Accuracy for fold {fold + 1}: {accuracy:.4f}")
    print(f"Converged in {epoch + 1} epochs\n")

# Output overall results:
average_accuracy = np.mean(accuracies)
average_iterations = np.mean(iterations_per_fold)   
print(f"\nAverage accuracy across all folds: {average_accuracy:.4f}")
print(f"Average number of epochs to converge: {average_iterations:.2f}")
```

```python
## TRANSFORM DATA USING WAVELET TRANSFORM ##
## Generate the 128 x 128 Haar wavelet matrix directly
haar_matrix = np.zeros((128, 128))

# First row of Haar matrix
haar_matrix[0, :] = 1 / np.sqrt(128)

# remaining rows in Haar matrix
n = 128
row = 1
length = n // 2

while length >= 1 and row < 128:
    for i in range(length):
        haar_matrix[row, 2 * i] = 1 
        haar_matrix[row, 2 * i + 1] = 1 
        if row + 1 < 128:
            haar_matrix[row + 1, 2 * i ] = 1 
            haar_matrix[row + 1, 2 * i + 1] = -1 
        row += 2
    length //= 2

# Normalize each row to become unit vector
for i in range(haar_matrix.shape[0]):
    row_norms = np.linalg.norm(haar_matrix[i])
    haar_matrix[i] = haar_matrix[i] / row_norms

H_128 = pd.DataFrame(haar_matrix).astype(np.float32)
print(H_128.head())
```

```python
# Create array to store the first four coefficients
wavelet_four_coeffs = np.zeros((df_norm_data.shape[0], 4)) 


for i in range(df_norm_data.shape[0]): # iterate through each row
    row = df_norm_data.iloc[i, :].values

    # Transformation: matrix multiplication
    transformed_row = np.dot(row, haar_matrix.T)

    # Store FIRST FOUR coefficients from transformed row
    wavelet_four_coeffs[i, :] = transformed_row[:4]

df_wavelet_coeffs = pd.DataFrame(wavelet_four_coeffs, columns=[f'Coeff_{i+1}' for i in range(4)]).astype(np.float32)
print("Shape of the wavelet tranformed dataset: ", df_wavelet_coeffs.shape)
print(df_wavelet_coeffs.head())
```
