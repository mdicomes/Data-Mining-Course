```python
import pandas as pd 
import numpy as np 
import time
```

```python
# Open .txt file using appropriate delimiter
df_data = pd.read_csv('Asgmnt1_data.txt', delimiter=',', header=None)

# Clean up the extra spaces and then split the columns
df_data_clean = df_data[0].str.strip().replace(r'\s+', ' ', regex=True)

# SPlit the columns using a single space as the delimiter
df_data_split = df_data_clean.str.split(' ', expand=True)

# Convert to numeric, coercing any non-convertible values to NaN
df_data_numeric = df_data_split.apply(pd.to_numeric, errors='coerce')


# Ensure datatypes are numeric
print(df_data_numeric.shape)
print(df_data_numeric.head())
```

```python
# Z-normalize across ROWS
df_norm_data = df_data_numeric.apply(lambda x: (x - x.mean()) / (x.std()), axis=1) # axis = 1 indicates row
print(df_norm_data.shape)
print(df_norm_data.head())
```

```python
## DISTANCE MATRIX USING EUCLIDEAN DISTANCE ##
# euclidean distance function #
def euclidean_distance(row1, row2):
    return np.sqrt(np.sum((row1 - row2) ** 2))
```

**Calculate Euclidean Distance on full, normalized dataset**

```python
## Calculate the Euclidean Distance on half of the normalized dataset ##

# collect time to calculate euclidean distance
start_time = time.time()

# Save the rows into an empty distance matrix
num_rows = df_norm_data.shape[0]
euclidean_distance_matrix = pd.DataFrame(np.zeros((num_rows, num_rows)))

# Euclidean Distance: Finding differences between samples in a pairwise fashion
for i in range(num_rows):
    for j in range(i, num_rows): # only computes for upper triangle (j >= i)
        dist = euclidean_distance(df_norm_data.iloc[i], df_norm_data.iloc[j])
        euclidean_distance_matrix.iat[i, j] = dist 
        euclidean_distance_matrix.iat[i, j] = dist # This is an error: resulted in no copying over of upper triangle. Should have been euclidean_distance_matrix.iat[j, i] = dist

end_time_df = time.time()

# Calculate elapsed time
elapsed_time_df = end_time_df - start_time

print(f'Time it takes to calculate euclidean distance on total dataframe: {elapsed_time_df/60} minutes')

print("First few values: ", euclidean_distance_matrix.head())
print("Last few values: ", euclidean_distance_matrix.tail())
```

```python
## Try block_size method when calculating the euclidean distance matrix ##
start_time = time.time()

# Save the rows into an empty distance matrix
num_rows = df_norm_data.shape[0]
block_size = 10000
euclidean_distance_matrix_2 = np.zeros((num_rows, num_rows), dtype=np.float32)

# Euclidean Distance: Finding differences between samples in a pairwise fashion
for i in range(0, num_rows, block_size):
    for j in range(i, num_rows, block_size): # only computes for upper triangle (j >= i)
        block_i = df_norm_data.iloc[i:i+block_size].to_numpy()
        block_j = df_norm_data.iloc[j:j+block_size].to_numpy()
        
        # pairwise distances between blocks
        for row_i in range(block_i.shape[0]):
            for row_j in range(block_j.shape[0]):
                dist = euclidean_distance(block_i[row_i], block_j[row_j])
                euclidean_distance_matrix_2[i+row_i, j+row_j] = dist 
                euclidean_distance_matrix_2[j+row_j, i+row_i] = dist

end_time_df = time.time()

# Calculate elapsed time
elapsed_time_df = end_time_df - start_time

euclidean_distance_matrix_2 = pd.DataFrame(euclidean_distance_matrix_2)

print(f'Time it takes to calculate euclidean distance on total dataframe: {elapsed_time_df/60} minutes')

print("First few values: ", euclidean_distance_matrix_2.head())
print("Last few values: ", euclidean_distance_matrix_2.tail())
```

__Wavelet Transform and Euclidean Distance__

```python
## TRANSFORM DATA USING WAVELET TRANSFORM ##
# USE THE FIRST FOUR COEFFICIENTS TO CALCULATE THE DISTANCE MATRIX

## Generate the 128 x 128 Haar wavelet matrix directly
haar_matrix = np.zeros((128, 128))

# First row of Haar matrix
haar_matrix[0, :] = 1 / np.sqrt(128)

# remaining rows in Haar matrix
n = 128
row = 1
length = n // 2

while length >= 1 and row < 128:
    for i in range(length):
        haar_matrix[row, 2 * i] = 1 
        haar_matrix[row, 2 * i + 1] = 1 
        if row + 1 < 128:
            haar_matrix[row + 1, 2 * i ] = 1 
            haar_matrix[row + 1, 2 * i + 1] = -1 
        row += 2
    length //= 2

# Normalize each row to become unit vector
for i in range(haar_matrix.shape[0]):
    row_norms = np.linalg.norm(haar_matrix[i])
    haar_matrix[i] = haar_matrix[i] / row_norms

H_128 = pd.DataFrame(haar_matrix).astype(np.float32)
print(H_128.head())

```

```python
## Verify rows are unit vectors and orthogonal ##
# Unit Vector #
row_unit_vector = np.linalg.norm(haar_matrix, axis=1)  # rows in matrix
print("unit vectors ", row_unit_vector)
print("Number of unit vectors (rows): ", len(row_unit_vector))

## Orthogonality ##
orth_vectors_wavelet = np.dot(haar_matrix[2], haar_matrix[3]) # checking orthogonality between first and second vector
print("Resulting dot product between second and third rows: ", orth_vectors_wavelet) # orth_vectors should be zero
```

```python
# Create array to store the first four coefficients
wavelet_four_coeffs = np.zeros((df_norm_data.shape[0], 4)) 

for i in range(df_norm_data.shape[0]): # iterate through each row
    row = df_norm_data.iloc[i, :].values

    # Transformation: matrix multiplication
    transformed_row = np.dot(row, haar_matrix.T)

    # Store first four coefficients from transformed row
    wavelet_four_coeffs[i, :] = transformed_row[:4]

df_wavelet_coeffs = pd.DataFrame(wavelet_four_coeffs, columns=[f'Coeff_{i+1}' for i in range(4)]).astype(np.float32)
print("Shape of the wavelet tranformed dataset: ", df_wavelet_coeffs.shape)
print(df_wavelet_coeffs.head())

```

```python
## Euclidean Distance for wt using level = 4 ##

start_time_WT = time.time()

# Save the rows into an empty distance matrix
num_rows = df_wavelet_coeffs.shape[0]
block_size = 1000
euclidean_distance_matrix_wt = np.zeros((num_rows, num_rows), dtype=np.float32)

# Euclidean Distance: Finding differences between samples in a pairwise fashion
for i in range(0, num_rows, block_size):
    for j in range(i, num_rows, block_size): # only computes for upper triangle (j >= i)
        block_i = df_wavelet_coeffs.iloc[i:i+block_size].to_numpy()
        block_j = df_wavelet_coeffs.iloc[j:j+block_size].to_numpy()
        
        # pairwise distances between blocks
        for row_i in range(block_i.shape[0]):
            for row_j in range(block_j.shape[0]):
                dist = euclidean_distance(block_i[row_i], block_j[row_j])
                euclidean_distance_matrix_wt[i+row_i, j+row_j] = dist 
                euclidean_distance_matrix_wt[j+row_j, i+row_i] = dist

end_time_WT = time.time()

# Calculate elapsed time
elapsed_time_WT = end_time_WT - start_time_WT

euclidean_distance_matrix_wt = pd.DataFrame(euclidean_distance_matrix_wt)

print(f'Time it takes to calculate euclidean distance on wt dataframe: {elapsed_time_WT/60} minutes')
print("First few values: ", euclidean_distance_matrix_wt.head())
print("Last few values: ", euclidean_distance_matrix_wt.tail())
```

**PCA, select best four PC, calculate Euclidean Distance**

```python
## Normalize original dataset per column ##
df_norm_col_data = df_norm_data.apply(lambda x: (x - x.mean()) / (x.std())).astype(np.float32) # df_data_numeric; float32 used due to memory constraints; default axis for .apply() is axis = 0 AKA columns

# number of samples (= number of rows)
n_samples_cov = df_norm_col_data.shape[0]

## Compute the covariance matrix ##
# Transpose normalized dataset #
df_norm_col_data_T = df_norm_col_data.T 
print("Shape of transposed dataframe: ", df_norm_col_data_T.shape)

cov_matrix = (df_norm_col_data_T.dot(df_norm_col_data) / (n_samples_cov - 1)).astype(np.float32) # this performs matrix multiplication between the normalized dataframe and its transpose, then divides by the number of samples - 1.
print("Shape of covariance matrix: ", cov_matrix.shape)
print(cov_matrix.head())
```

```python
## Eigenvalue decomposition ##
eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix) # eigenvalues: 1D array, eigenvectors: 2D array; .eigh used because covariance matrix is symmertric.

## Verify eigenvectors are unit vectors and orthogonal ##
# Unit Vector #
mag_vector = np.linalg.norm(eigenvectors[:, 1]) # this will find the magnitude of the second eigenvector
print("magnitude of second eigenvector: ", mag_vector)

## Orthogonality ##
orth_vectors = np.dot(eigenvectors[:, 0], eigenvectors[:, 1]) # checking orthogonality between first and second vector
print("Resulting dot product between first and second vectors: ", orth_vectors) # orth_vectors should be zero
```

```python
## Sort eigenvectors by largest to smallest eigenvalues ##
sorted_indices = np.argsort(eigenvalues)[::-1] # returns the indices that will sort the eigenvalues in descending order
sort_eigenvalues = eigenvalues[sorted_indices] # sorts eigenvalues based on the above indices
sort_eigenvectors = eigenvectors[:, sorted_indices] # sorts eigenvectors based on the above indices

## Select top four principle components ##
top_eigenvalues = sort_eigenvalues[:4]
top_eigenvectors = sort_eigenvectors[:, :4]
print("Length of top eigenvalues :", len(top_eigenvectors))
print("Top Eigenvalues: ", top_eigenvalues)
print("Top Eigenvectors: ", top_eigenvectors)
```

```python
## Transform the Data ##
transformed_data = np.dot(df_norm_col_data, top_eigenvectors)

# DataFrame
df_pca = pd.DataFrame(transformed_data, columns=[f'PC{i+1}' for i in range(4)]).astype(np.float32)
print("Shape of transformed data using top 4 eigenvectors: ", df_pca.shape)
print(df_pca.head())
```

```python
## Euclidean Distance ##
start_time_PCA = time.time()

# Save the rows into an empty distance matrix
num_rows = df_pca.shape[0]
block_size = 10000
euclidean_distance_matrix_pca = np.zeros((num_rows, num_rows), dtype=np.float32)

# Euclidean Distance: Finding differences between samples in a pairwise fashion
for i in range(0, num_rows, block_size):
    for j in range(i, num_rows, block_size): # only computes for lower triangle (j >= i)
        block_i = df_pca.iloc[i:i+block_size].to_numpy()
        block_j = df_pca.iloc[j:j+block_size].to_numpy()
        
        # pairwise distances between blocks
        for row_i in range(block_i.shape[0]):
            for row_j in range(block_j.shape[0]):
                dist = euclidean_distance(block_i[row_i], block_j[row_j])
                euclidean_distance_matrix_pca[i+row_i, j+row_j] = dist 
                euclidean_distance_matrix_pca[j+row_j, i+row_i] = dist

end_time_PCA = time.time()

# Convert to DataFrame
euclidean_distance_matrix_pca = pd.DataFrame(euclidean_distance_matrix_pca)

# Calculate elapsed time
elapsed_time_WT = end_time_PCA - start_time_PCA

print(f'Time it takes to calculate euclidean distance on pca data: {elapsed_time_WT/60} minutes')

print("First few values: ", euclidean_distance_matrix_pca.head())
print("Last few values: ", euclidean_distance_matrix_pca.tail())
```

```python
## Randomly selecting pairs to calculate percentage of matching relationships ##
import random

euclidean_distance_matrix_np = np.array(euclidean_distance_matrix_2)
euclidean_distance_matrix_pca_np = np.array(euclidean_distance_matrix_pca)

# number of random pairs
n_samples = 5000

# randomly sample pairs of indices from the distance matrix
num_rows = euclidean_distance_matrix_np.shape[0]
random_pairs = [(random.randint(0, num_rows-1), random.randint(0, num_rows-1)) for _ in range(n_samples)]

# initial counters 
total_pairs = 0
pca_matching_relationships = 0

# Loop through the sampled pairs of cells 
for (i, j) in random_pairs:
    if i != j: # don't compare distances between same points
        # Ground truth distances
        dist_gt_1 = euclidean_distance_matrix_np[i, j]

        # Another random pair
        random_pair2 = (random.randint(0, num_rows - 1), random.randint(0, num_rows - 1))
        dist_gt_2 = euclidean_distance_matrix_np[random_pair2[0], random_pair2[1]]

        # Check relationship in ground truth matrix
        gt_relationship = dist_gt_1 > dist_gt_2

        # Distances from PCA distance matrix
        dist_pca_1 = euclidean_distance_matrix_pca_np[i, j]
        dist_pca_2 = euclidean_distance_matrix_pca_np[random_pair2[0], random_pair2[1]]

        # Check relationship in pca distance matrix
        pca_relationship = dist_pca_1 > dist_pca_2

        # Compare PCA relationship with ground truth relationship
        if pca_relationship == gt_relationship:
            pca_matching_relationships += 1

        total_pairs += 1

# Calculate the percentage of matching relationships
percentage_matching = (pca_matching_relationships / total_pairs) * 100

print(f"Percentage of matching relationships in PCA: {percentage_matching:.2f}%")
```
